```python
from tqdm import tqdm  # Impor modul tqdm

# Baca file "nama.txt" dan simpan nama-nama dalam sebuah set
nama_set = set()
with open("nama.txt", "r") as file:
    for line in file:
        nama = line.strip()
        if nama:
            nama_set.add(nama)

# Fungsi untuk membersihkan teks dalam kolom "jawab" berdasarkan patokan nama
def clean_text_based_on_names(text):
    for nama in nama_set:
        # Mencari dan menghapus setiap kemunculan nama dalam teks
        text = re.sub(r'\b{}\b'.format(re.escape(nama)), '', text, flags=re.IGNORECASE)
    
    # Menghapus karakter non-ASCII dan mengganti spasi ganda dengan satu spasi
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# Contoh penggunaan dengan tqdm untuk melihat bar proses
# DataFrame Anda disimpan dalam variabel df
tqdm.pandas()  # Menggunakan metode .progress_apply() dari pandas dengan tqdm
df['jawab'] = df['jawab'].progress_apply(clean_text_based_on_names)
```
