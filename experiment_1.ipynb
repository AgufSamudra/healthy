```python
import pandas as pd
import numpy as np
import re
# from Sastrawi.Stemmer.StemmerFactory import StemmerFactory # stemming
import time
from tqdm import tqdm
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
import logging
from gensim.models import FastText

import os
# nltk.download('punkt') # Tokenize
```

```python
df = pd.read_csv("dataset/data_13_labels.csv", skipinitialspace=True)
# df.drop(columns="Unnamed: 0", axis=1, inplace=True)
df.head()
```

```python
def preprocessing(text):
    text =  re.sub(r'[\n+]', '', str(text))
    text =  re.sub(r'[^\w\s]', '', text)
    text =  re.sub(r'[^\x00-\x7F]+', '', text)
    text =  re.sub(r'(pagi|siang|sore)', '', text, flags=re.IGNORECASE)
    text =  re.sub(r'(dok|dokter)', '', text, flags=re.IGNORECASE)
    text =  re.sub(r'(terimakasih|terima kasih|alo|alodokter)', '', text, flags=re.IGNORECASE)
    text =  re.sub(r'\s+', ' ', text).strip().lower()
    text = word_tokenize(text)
    
    return text
```

```python
sentence = [preprocessing(text_prep) for text_prep in tqdm(df.tanya, leave=False)]
```

```python
model = FastText(sentence, vector_size=100, workers=-1, epochs=50)
```

```python
model.save("language_model/healthy_chat_13_labels.bin")
```

# Continue Training

```python
model = FastText.load("language_model/healthy_chat_13_labels.bin")
```

```python
model.train(sentence, total_examples=len(sentence), epochs=1)
```

```python
model.save("model_train//healthy_chat.bin")
```
